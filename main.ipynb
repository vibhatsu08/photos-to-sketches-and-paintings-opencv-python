```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# image = cv2.imread("./images/image-1.jpg")
# image = cv2.imread("./images/image-2.jpg")
image = cv2.imread("./images/image-3.jpg")

plt.imshow(image)
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_grayscale = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)
plt.imshow(image_grayscale, cmap="gray")
```

```python
image_blur = cv2.GaussianBlur(image_grayscale, (5,5), 5)
plt.imshow(image_blur, cmap='gray')
```

```python
sobelx = cv2.Sobel(image_blur, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(image_blur, cv2.CV_64F, 0, 1, ksize = 5)
gradmag_square = np.square(sobelx) + np.square(sobely)
gradmag = np.sqrt(gradmag_square)
plt.imshow(gradmag, cmap = "gray")
```

```python
gradmag_inverse = 255 - gradmag
plt.imshow(gradmag_inverse, cmap="gray")
```

```python
threshold_value, threshold_image = cv2.threshold(gradmag_inverse, 10, 255, cv2.THRESH_BINARY)
plt.imshow(threshold_image, cmap="gray")
```

```python
pencilsketch_gray, pencilsketch_color = cv2.pencilSketch(image_rgb, sigma_s=60, sigma_r=0.05, shade_factor=0.05)
plt.imshow(pencilsketch_gray, cmap="gray")
```

```python
plt.imshow(pencilsketch_color)
```

```python
waterpaint_image = cv2.stylization(image_rgb, sigma_s=100, sigma_r=0.08)
plt.imshow(waterpaint_image)
```
